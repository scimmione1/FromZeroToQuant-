```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
# import the path
path = os.path.abspath(os.path.join("..", "01_get_the_data", "spy_historical_data.csv"))
```

```python
# Load the CSV
df = pd.read_csv(path)

# Drop any rows that are completely NaN (e.g. 'Date' row)
df = df.dropna(how='all')

# convert df.datetime to datetime
df['datetime'] = pd.to_datetime(df['datetime'], format='%Y-%m-%d %H:%M:%S')

# Show the result
df.head()
```

```python
# Calculate simple returns with pct_change()
simple_returns = df['SPY'].pct_change().fillna(0)

daily_returns = simple_returns.copy()
```

```python
# plot daily_returns and datetime
fig = plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], daily_returns, label='Daily Returns')
plt.title('Daily Returns of SPY')
plt.xlabel('Time')
plt.ylabel('Daily Return')
plt.legend()
plt.grid()
plt.show()
```
