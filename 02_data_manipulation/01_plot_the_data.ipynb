```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
# import the path
path = os.path.abspath(os.path.join("..", "01_get_the_data", "spy_historical_data.csv"))
```

```python
# Load the CSV
df = pd.read_csv(path)

# Drop any rows that are completely NaN (e.g. 'Date' row)
df = df.dropna(how='all')

# convert df.datetime to datetime
df['datetime'] = pd.to_datetime(df['datetime'], format='%Y-%m-%d %H:%M:%S')

# Show the result
df.head()
```

```python
# plot SPY and datetime columns
plt.figure(figsize=(12,6))
plt.plot(df['datetime'], df['SPY'], label='SPY Close Price')
plt.title('SPY Historical Close Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```
