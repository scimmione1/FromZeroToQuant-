```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
# Costruisci il path relativo al file CSV rispetto al notebook
path = os.path.abspath(os.path.join("..", "01_get_the_data", "spy_historical_data.csv"))
```

```python
# Load the CSV with MultiIndex columns (Tickers, OHLCV)
df = pd.read_csv(path)

# Drop any rows that are completely NaN (e.g. 'Date' row)
df = df.dropna(how='all')

# Show the result
df.head()
```

```python
# plot SPY column
plt.figure(figsize=(12,6))
plt.plot(df['SPY'], label='SPY Close Price')
plt.title('SPY Historical Close Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```
