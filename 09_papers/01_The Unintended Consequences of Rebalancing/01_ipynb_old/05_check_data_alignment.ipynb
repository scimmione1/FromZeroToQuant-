```python
import nbformat
import pandas as pd
from io import StringIO

# === CONFIG ===
file1_path = "01_The Unintended Consequences of Rebalancing.ipynb"
file2_path = "02_The Unintended Consequences of Rebalancing_xl.ipynb"
file3_path = "03_The Unintended Consequences of Rebalancing_tv.ipynb"

# === Helper function to extract DataFrames from a notebook ===
def extract_dataframes_from_notebook(nb_path):
    """
    Reads a .ipynb file, finds code cells that define pandas DataFrames,
    and returns a dictionary {variable_name: DataFrame}.
    """
    with open(nb_path, 'r', encoding='utf-8') as f:
        nb = nbformat.read(f, as_version=4)
    
    dataframes = {}
    namespace = {}
    for cell in nb['cells']:
        if cell['cell_type'] == 'code':
            try:
                exec(cell['source'], namespace)
            except Exception:
                # Skip cells that cannot be executed (missing imports, etc.)
                continue
    # Extract only pandas DataFrames
    for k, v in namespace.items():
        if isinstance(v, pd.DataFrame):
            dataframes[k] = v
    return dataframes

# === Step 1: Extract DataFrames from all three notebooks ===
dfs1 = extract_dataframes_from_notebook(file1_path)
dfs2 = extract_dataframes_from_notebook(file2_path)
dfs3 = extract_dataframes_from_notebook(file3_path)

print(f"Extracted {len(dfs1)} DataFrames from File 1")
print(f"Extracted {len(dfs2)} DataFrames from File 2")
print(f"Extracted {len(dfs3)} DataFrames from File 3")

# === Step 2: Identify date columns ===
def get_date_column(df):
    """Return the most likely date column or index."""
    for col in df.columns:
        if pd.api.types.is_datetime64_any_dtype(df[col]) or "date" in col.lower():
            return col
    if pd.api.types.is_datetime64_any_dtype(df.index):
        return df.index.name
    return None

# Choose one main DataFrame per notebook (first one found)
df1 = list(dfs1.values())[0]
df2 = list(dfs2.values())[0]
df3 = list(dfs3.values())[0]

# Get date columns
col1 = get_date_column(df1)
col2 = get_date_column(df2)
col3 = get_date_column(df3)

if col1 and col2 and col3:
    d1 = df1[col1] if col1 in df1.columns else df1.index
    d2 = df2[col2] if col2 in df2.columns else df2.index
    d3 = df3[col3] if col3 in df3.columns else df3.index
else:
    raise ValueError("Could not detect date columns in one or more files.")

# === Step 3: Find alignment differences ===
def compare_dates(d_ref, d_test, label_ref, label_test):
    d_ref = pd.to_datetime(d_ref)
    d_test = pd.to_datetime(d_test)
    missing_in_test = sorted(set(d_ref) - set(d_test))
    extra_in_test = sorted(set(d_test) - set(d_ref))
    print(f"\n Alignment check: {label_test} vs {label_ref}")
    print(f"Missing in {label_test}: {len(missing_in_test)} dates")
    print(f"Extra in {label_test}: {len(extra_in_test)} dates")
    if missing_in_test[:5]:
        print(" First missing dates:", missing_in_test[:5])
    if extra_in_test[:5]:
        print(" First extra dates:", extra_in_test[:5])

# Compare File 1 and 3 against File 2
compare_dates(d2, d1, "File 2", "File 1")
compare_dates(d2, d3, "File 2", "File 3")

print("\n Data alignment analysis complete.")
```
